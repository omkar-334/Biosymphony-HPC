```python
import numpy as np

def read_xyz(fname:str):

    path = fname

    with open(path, 'r') as file:

        numAtoms = file.readline()
        sysName = file.readline()

        system_points = np.empty((0,2))

        for i in range(int(numAtoms)):

            line = file.readline()

            atName1, x1, y1, z1 = line.split()

            system_points = np.append(system_points, np.array([[x1, y1]]), axis=0)

    return system_points

system_points = read_xyz('Ar_1681_50_50_2D.xyz')

system_points
```

```python
all_neighbours = []

def find_neighbours(system_points, query_points, box_size):

  box_size = [50,50]

  cutoff_distance = 5

  kdtree = scipy.spatial.KDTree(system_points, boxsize=box_size)

  distances, neighbours = kdtree.query(query_points, k=100, distance_upper_bound=cutoff_distance)

  all_neighbours.append( list(neighbours)*5000)

  return neighbours, distances
```

```python
import tracemalloc
```

```python
tracemalloc.start()

neighbours, distances = find_neighbours(system_points, query_points, box_size)

snapshot = tracemalloc.take_snapshot()

top_stats = snapshot.statistics('traceback')

top_stats[0]
```
